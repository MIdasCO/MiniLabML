````python
import pandas as pd
import sqlite3
````

### 2.	Create SQLite Database and Table
	•	Connect to an SQLite database using sqlite3.
	•	Define a table with appropriate columns for each attribute in the dataset.
	•	Use SQL data types that best match the attributes (e.g., TEXT for Series_Title, INTEGER for Released_Year).

````python
conn = sqlite3.connect("imdb_top_1000.db")
cursor = conn.cursor()
````

````python
df = pd.read_csv("imdb_top_1000.csv")
df.head(1)
````

````python
cursor.execute("""
    CREATE TABLE imdb(
    id_imbd INT PRIMARY KEY,
    Poster_Link	TEXT,
    Series_Title TEXT,
    Released_Year TEXT,
    Certificate TEXT,
    Runtime TEXT,
    Genre TEXT,
    IMDB_Rating FLOAT,
    Overview TEXT,
    Meta_score FLOAT,
    Director TEXT,
    Star1 TEXT,
    Star2 TEXT,
    Star3 TEXT,
    Star4 TEXT,
    No_of_Votes INT,
    Gross TEXT
    )
""")
````

### 3.	Insert Data from DataFrame to SQLite
	•	Use df.to_sql() to insert data directly from the DataFrame into the SQLite table.

````python
df['Released_Year'].value_counts()
````

````python
df.to_sql('imdb', conn, if_exists='replace', index=False)
````

## Part 2. Writing SQL Queries (0.75 p)

### 1. Select specific columns: Retrieve only Series_Title, IMDB_Rating, and Genre for each movie released after 2010.

````python
cursor.execute("""
    select Series_Title,IMDB_Rating,Genre
    from imdb
    where Released_Year>2010
""").fetchall()
````

### 2. Top-rated movies: Select movies with IMDB_Rating above 8.5.

````python
cursor.execute("""
    select *
    from imbd
    where IMDB_Rating>8.5
""").fetchall()
````

### 3.	Aggregate and Group By Queries: Count by genre: Count how many movies are available per genre.


````python
cursor.execute("""
    SELECT Genre, COUNT(*)
    FROM imdb
    GROUP BY Genre;
""").fetchall()
````
